<h1><font color="darkblue">Hypothesis Testing</font></h1>

### Why we need hypothesis testing
- Usually, data is collected with the intent of proving or disproving a hypothesis

    - $H_{0}$: Null hypothesis - one dataset is sampled from another
    - $H_{A}$: Alternative hypothesis - the two datasets are distinct
- This is done by compiling two datasets (e.g., a population and a sample dataset)
- t-test allows us to draw conclusions about which hypotheses we should reject/accept

**Goal of a t-test: quantify statistical significance**

### How do we quantify the difference between the population and sample means? 

Assume the data follows a normal distribution. 

$\mu$: Population mean (i.e., the true mean)

$\bar{X}$: Sample mean

$\bar{X}-\mu$: How much the sample distribution average differs from the population

- How do we account for the standard deviation? 

### t-statistic
>$t = \frac{\bar{X}-\mu}{\sigma_{\bar{X}}}$

$\sigma_{\bar{X}}$: Standard error of sample mean - how much a given distribution deviates from the population mean

- The variance of a sample distribution means is related to the population distribution by

>$\sigma^{2}_{\bar{X}} = \sigma^{2}/{n}$

$\sigma$: Population standard deviation

$n$: size of sample datasets

Thus, 

>$\sigma_{\bar{X}} = \sigma/\sqrt{n}$



- Approximate $\sigma\approx s$. Also, assume sample values are independent. 

>$\sigma_{\bar{X}} = s/\sqrt{n}$

$s$: Sample standard deviation

- The t-value is then

>$t = \frac{\bar{X}-\mu}{s/\sqrt{n}}$

**Note: not the same as a z-statistic!**


### p-value
Now that we have the number of standard deviations we are from the population mean, we then want to convert this into a probability. 


>$p$: p-value - the probability that we would get the sample mean assuming null hypothesis to be true

To obtain $p$, we must insert $t$ into the t-distribution (not the same as a normal distribution). 

If $p$ is sufficiently low, we must reject the null hypothesis, $H_{0}$

### Bayes' Rule

> Given two discrete events: $X$ and $Y$, 
>
>$\displaystyle P(X|Y) = \frac{P(X)\,P(Y|X)}{P(Y)}$ 
>
- $P(X|Y)$: Posterior - the probability of $X$ given $Y$
>
- $P(X)$: Prior - the probability of $X$ prior to obtaining any data
>
- $P(Y|X)$: Likelihood - the probability of $Y$ given $X$
>

### Also works for continuous distributions

>$\displaystyle p(x|y) = \frac{p(x)\,p(y|x)}{p(y)}$ 
>
>Or...
>
>$\displaystyle p(\theta|D) = \frac{\pi(\theta)\,{\cal{}L}_D(\theta)}{Z}$ 
>
> Where: 
- $\theta$: Parameters of a model
- $D$: Data
- $\pi(\theta)$: Prior
- ${\cal{}L}_D(\theta)$: Lilkelihood function
- $Z$: Normalization constant (to make $\displaystyle \int p(\theta|D)\,dx = 1$)

### How to feed your model new data
For $N$ independent measurements: 
>${\cal{}L}_D(\theta) = \prod_{i=1}^N\ell_{i}(\theta)$

- Collect as many data points ($N$) as possible. 
- If new data is obtained, update the likelihood function (using above equation). 
- Calculate new and improved posterior. 

>$\displaystyle p(\theta|D) \propto \big[ Prior] \big[ Likelihood]$ 

Initial posterior:

>$\displaystyle p(\theta|D) \propto \big[ \pi(\theta)] \big[ \ell_{1} \big]$ 

After an update, 

>$\displaystyle p(\theta|D) \propto \big[ \pi(\theta) \ell_{1} \big] \big[ \ell_{2} \big]$ 

After another update, 
>$\displaystyle p(\theta|D) \propto \big[ \pi(\theta) \ell_{1} \ell_{2} \big] \big[ \ell_{3} \big]$ 

And so on...

**In other words: "Today's posterior is tomorrow's prior."**   - Dennis Lindley

### Bayesian inference example 

We wish to measure the probability that a coin will yield a heads if flipped. We do not know if the coin is fair. Bayesian inference can help us predict that probability distribution of the coin. 

Use a Bernoulli distribution:

$\ell_{i}(n_{i}, s_{i}) = {{n_{i}}\choose{s_{i}}} p^{s_{i}}(1-p)^{n_{i}-s_{i}}$


$\displaystyle p(\theta|D) \propto \pi(\theta)\,{\cal{}L}_D(\theta)$ 

We can run a number of trials to update the likelihood function with new data like so, 

$\displaystyle p(\theta|D) \propto \pi(\theta) \bigg[p^{s_{1}}(1-p)^{n_{1}-s_{1}} \times p^{s_{2}}(1-p)^{n_{2}-s_{2}}  \times \dots  \bigg]$ 

This can be simplified as

$\displaystyle p(\theta|D) \propto \pi(\theta) p^{S}(1-p)^{N-S}$ 

Now, assume a flat prior (i.e., $\pi(\theta)=1$)

$\displaystyle p(\theta|D) \propto p^{S}(1-p)^{N-S}$ 

>$N$: Total number of trials
>
>$S$: Total number of heads

### Conjugate Priors
A conjugate prior is a prior in which the corresponding posterior has the same type of distribution. This makes updating the posterior relatively simple, mathematically. Also, this gives a closed-form analytical distribution for the posterior. 

Example: Prove that a normal distribution is its own conjugate. 

>${\cal{}L}_D(\mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}}\ e^{-\frac{(p-\mu)^2}{2\sigma^2}}$

>$\pi(\theta) = \frac{1}{\sqrt{2\pi s^2}}\ e^{-\frac{(p-m)^2}{2s^2}}$

Multiplying the distributions, 

>$\pi(\theta)\,{\cal{}L}_D(\theta) = C e^{-\frac{(p-\mu)^2}{2\sigma^2}} e^{-\frac{(p-m)^2}{2s^2}}$

>$\pi(\theta)\,{\cal{}L}_D(\theta) = C e^{-\frac{(p-\mu)^2}{2\sigma^2}-\frac{(p-m)^2}{2s^2}} $

Defining:

>$\frac{1}{\sigma_n^2} = \frac{1}{\sigma^2} + \frac{1}{s^2}$

>$\mu_n = \sigma_n^2\bigg(\frac{m}{s^2}+\frac{\mu}{\sigma^2}\bigg)$


>$\pi(\theta)\,{\cal{}L}_D(\theta) \propto e^{-\frac{(p-\mu_n)^2}{2\sigma_n^2}}$

This is a normal distribution form. Thus, if a likelihood function has a normal distribution, using a normal prior will ensure that the posterior is also normal. 

### Application to hypothesis testing
Bayesian inference can also be applied to hypothesis testing. 

- Say the null hypothesis, $H_{0}$, is represented by the model $M_{1}$. 

- Also, the alternative hypothesis, $H_{A}$, is represented by $M_{2}$. 

From Bayes' rule, 

$P(M_{1}|data) = \frac{P(M_{1})\,P(data|M_{1})}{P(data)} \quad$     $\quad P(M_{2}|data) = \frac{P(M_{2})\,P(data|M_{2})}{P(data)}$

We define a Bayes' factor:
$K = \frac{P(data|M_{1})}{P(data|M_{2})}$
which is a measure of the relative strength of the models. 
- If $K$ is large, the null hypothesis is likely to be true
- If $K$ is small, the null hypothesis should likely be rejected

Assuming we know the model distribution, we can apply a given sample data to both models to calculate the Bayes factor.
